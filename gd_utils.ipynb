开始拷贝，**别忘了配置nodejs环境，上面有一键配置。**

````
import os
import time
fileid = input("文件ID：")
destid = input("目标ID：")
starttime = time.time()
print("开始拷贝...")
os.system("cd /content/gd-utils && node --max-old-space-size=8192 copy " + fileid + " " + destid + " -S > copy.txt")
with open("copy.txt") as fp:
  o = fp.read()
print(o)
endtime = time.time()
timeelapsed = endtime - starttime
print("拷贝完成！")
print("用时：" + str(timeelapsed) + "秒")
````

统计

````
import os
import time
import json
fileid = input("文件ID：")
starttime = time.time()
print("开始统计...")
os.system("node --max-old-space-size=8192 /content/gd-utils/count " + fileid + " -S -u > stat.txt")
with open("stat.txt") as fp:
  o = fp.read()
print(o)
endtime = time.time()
timeelapsed = endtime - starttime
print("统计完成！")
print("用时：" + str(timeelapsed) + "秒")
````
